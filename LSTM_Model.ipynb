```python
import os
import pandas as pd

# Create a new dataframe for coombining all exeercises
newdf = pd.DataFrame()
print(newdf)

# Read the csv file for each exercise

for i in os.scandir('/content/drive/MyDrive/Pose_Estimation/Mediapipe_featurecsv'):
  df = pd.read_csv(i.path) # read current csv into df
  print(len(df))
  # df = df.assign(Exercise_name = i.path.split("/")[6].split(".")[0])
  newdf = pd.concat([newdf,df],axis=0) # add current df to new df
  print("New length: ",len(newdf))
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
newdf['Exercise_name']= label_encoder.fit_transform(newdf['Exercise_name'])
```

```python
label_encoder.classes_
```

```python
from sklearn.model_selection import train_test_split

X = newdf.drop(['Exercise_name'],axis=1)
Y = newdf['Exercise_name']

X = (X-X.mean(axis=0))/X.std(axis=0)
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.30)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from tensorflow.keras.utils import to_categorical

y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
from tensorflow.keras.preprocessing import timeseries_dataset_from_array

ds = timeseries_dataset_from_array(
     data= X_train,
     targets= y_train,
     sequence_length=30,
     batch_size = 2300)
```

**LSTM**

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.layers import Dropout,BatchNormalization
```

```python
model = Sequential()
model.add(LSTM(1024, return_sequences=True, activation='relu', input_shape=(30, 99)))
model.add(Dropout(0.2))
model.add(LSTM(512, return_sequences=True, activation='relu'))
model.add(Dropout(0.2))
model.add(LSTM(256, return_sequences=False, activation='relu'))
model.add(BatchNormalization())
model.add(Dense(256, activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(10, activation='softmax'))
model.summary()
```

```python
model.compile(optimizer='Adam', loss='categorical_crossentropy', metrics=['categorical_accuracy'])
```

```python
model.fit(x,y,validation_data=(xval, yval), epochs=50, batch_size=1)
```

```python
import math
import cv2
import numpy as np
from time import time
import mediapipe as mp
import matplotlib.pyplot as plt
```

```python
mp_pose = mp.solutions.pose

# Setting up the Pose function.
pose = mp_pose.Pose(static_image_mode=True, min_detection_confidence=0.3, model_complexity=2)

# Initializing mediapipe drawing class, useful for annotation.
mp_drawing = mp.solutions.drawing_utils
```

```python
keypoint_list = []
for i in range(33):
  keypoint_list.append(mp_pose.PoseLandmark(i).name.lower())
```

```python
column_names = []
for i in keypoint_list:
  column_names.append(i+"_x")
  column_names.append(i+"_y")
  column_names.append(i+"_z")
```

```python
import pandas as pd

test_df = pd.DataFrame(columns = column_names)
test_df
```

```python
with mp_pose.Pose(static_image_mode=False, min_detection_confidence=0.5, model_complexity=1)as pose:
  # for vd in video_paths:
    video = cv2.VideoCapture("/content/data/000000.mp4")
    frame_count = -1
    while video.isOpened():
      # Read a frame.
      ok, frame = video.read()

      # Check if frame is not read properly.
      if not ok:

          # Break the loop.
          break

      frame_count = frame_count+1

      if frame_count%3!=0:
        continue

      feature_list = []
      # feature_list.append('Superman')
      # feature_list.append(frame_count)
      results = pose.process(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
      for i in range(33):
        try:
          vals = results.pose_landmarks.landmark[mp_pose.PoseLandmark(i).value]
          #print(mp_pose.PoseLandmark(i).name)
          #print(vals)
          x = vals.x
          y = vals.y
          z = vals.z
          #print(x,y,z)
          feature_list.append(x)
          feature_list.append(y)
          feature_list.append(z)

        except:
         continue
```

```python
import numpy as np

test = list(test_df.iloc[0])
test_arr = np.array(test)
test_df.loc[len(test_df)] = feature_list
```

```python
test_arr = (test_arr-test_arr.mean())/test_arr.std()
```

```python
res = model.predict(test_arr.reshape(1,99))
```
